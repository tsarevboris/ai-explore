```python
from fastai.vision.all import *
from fastai.vision.widgets import *
from io import BytesIO
```

```python
learn_inf = load_learner('../models/coffee.pkl')
```

```python
btn_upload = widgets.FileUpload()
btn_run = widgets.Button(description='Classify')
out_pl = widgets.Output()
lbl_pred = widgets.Label()

def on_click_classify(change):
    # Check if file is uploaded
    if not btn_upload.value:
        lbl_pred.value = 'Please upload an image first!'
        return
    
    # Get the first uploaded file
    uploaded_file = btn_upload.value[0]
    
    # Wrap the bytes in BytesIO so PIL can read itg
    img = PILImage.create(BytesIO(uploaded_file['content']))
    
    out_pl.clear_output()
    with out_pl:
        display(img.to_thumb(128,128))

    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)

VBox([widgets.Label('Select your coffee!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
