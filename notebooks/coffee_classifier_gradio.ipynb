```python
from fastai.vision.all import *
```

```python
learn = load_learner('../models/coffee.pkl')
```

```python
labels = learn.dls.vocab

def predict(img):
    img = PILImage.create(img)
    pred, pred_idx, probs = learn.predict(img)
    return {labels[i]: float(probs[i]) for i in range(len(labels))}
```

```python
import gradio as gr

gr.Interface(
    title = "Coffee Classifier",
    description = "Classifies coffee images between espresso, americano, capuccino and latte"
    fn=predict,
    inputs=gr.Image(type="pil"),
    outputs=gr.Label(num_top_classes=3),
    examples = ['images/coffee capuccino.jpg', 'images/coffee espresso.jpg']
).launch(share=True)
```
