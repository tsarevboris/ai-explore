## Python Basics

### Binding
By default assignment (also called *binding*) creates *the reference*  

```python
a = [1, 2, 3, 4]
b = a

a.append(5)
b # [1, 2, 3, 4, 5]
```

Same for function arguments

```python
def append_to_list(l, value):
    l.append(value)

a = [1, 2, 3, 4]
append_to_list(a, 5)
a
```

```python
def append_and_copy(l, value):
    new_l = l.copy()
    new_l.append(value)
    return new_l

a = [1, 2, 3, 4]
b = append_and_copy(a, 5)
a, b
```

```python
a = [1, 2, 3, 4]
b = a
c = list(a)
a is b, a is c
```
